```python
#Cambio el type de last_update de object a datetime

rental['last_update'] = pd.to_datetime(rental['last_update'])
```

```python
#Cambio el type de rental_date de object a datetime

rental['rental_date'] = pd.to_datetime(rental['rental_date'])
```

```python
#Cambio el type de return_date de object a datetime

rental['return_date'] = pd.to_datetime(rental['return_date'])
```

```python
#Voy a crear una columna nueva que sea el numero de dias que rento la pelicula

rental['dias_renta'] = round((rental['return_date'] - rental['rental_date']).dt.total_seconds() / (60 * 60 * 24)) # Diferencia en días

rental
```

```python
#Tome unicamente los 1000 inventory_id 
rental = rental[rental['inventory_id']<=1000]
rental
```
